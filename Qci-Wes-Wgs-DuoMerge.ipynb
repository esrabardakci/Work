Duo_Merge function takes 2 input files and output file name and writes merged version of two input files in a excel file which is named as output file name given at the beginning.

```python
def Duo_Merge(file1_path, file2_path,output_filename, index1, index2): 
    data1=pd.read_excel(file1_path)
    data2=pd.read_excel(file2_path)
    #string1='MG25-3947'
    #string2='MG25-3948'
    data_duo=pd.merge(data1,data2,on=['Chromosome','Position','End Position','Reference Allele','Sample Allele'],how='outer',suffixes=('*'+index1,'*'+index2))
    cols=sorted(data_duo.columns,key=lambda x: data1.columns.get_loc(x.split('*')[0]))
    data_duo_update=data_duo[cols]
    data_duo_update.columns= data_duo_update.columns.str.replace(r"*","_")
    data_duo_update.to_excel(output_filename,index=False)
```

```python
#DEĞİŞTİR
Duo_Merge('MG25-4143.xlsx','MG25-4144.xlsx','MG25-4143-4144-WGS-DUO.xlsx', index1, index2)
```

```python
def merge_specified_columns(df, columns):
    def merge_row(row):
        values = [row[col] for col in columns if pd.notna(row[col]) and row[col] != '']
        if not values:
            return None
        if all(value == values[0] for value in values):
            return values[0]
        return values[0]  # If not identical, return the first non-empty value (or define another rule if needed)
    
    return df.apply(merge_row, axis=1)
```
