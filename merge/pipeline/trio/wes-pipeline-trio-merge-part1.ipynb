```python
import pandas as pd
import re
```

```python
file1_path='W1097347_results.xlsx'
file2_path='W1097346_results.xlsx'
file3_path='W1087820_results.xlsx'
```

```python
data1=pd.read_excel(file1_path)
data2=pd.read_excel(file2_path)
data3=pd.read_excel(file3_path)
```

```python
string1=re.findall(pattern='\d+(?=_)',string=file1_path)[0]
string2=re.findall(pattern='\d+(?=_)',string=file2_path)[0]
string3=re.findall(pattern='\d+(?=_)',string=file3_path)[0]
```

```python
string3
```

```python
cols=[a for a in data3.columns if a not in  ['#Chr', 'Start', 'End', 'Ref', 'Alt']]
a='*'+string3
```

```python
a
```

```python
data3.columns = ['{}{}'.format(c, str(a) if c in cols else '') for c in data3.columns]
```

```python
data3
```

```python
data_duo=pd.merge(data1,data2, on=['#Chr','Start','End','Ref','Alt'],how='outer',suffixes=('*'+string1,'*'+string2))
```

```python
data_duo
```

```python
data_trio=pd.merge(data_duo,data3,on=['#Chr','Start','End','Ref','Alt'],how='outer')
```

```python
data_trio
```

```python
col=sorted(data_trio.columns,key=lambda x: data1.columns.get_loc(x.split('*')[0]))
```

```python
data_trio_update=data_trio[col]
```
