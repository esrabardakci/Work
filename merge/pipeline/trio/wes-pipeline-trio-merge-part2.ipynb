```python
def merge_specified_columns(df, columns):
    def merge_row(row):
        values = [row[col] for col in columns if pd.notna(row[col]) and row[col] != '']
        if not values:
            return None
        if all(value == values[0] for value in values):
            return values[0]
        return values[0]  # If not identical, return the first non-empty value (or define another rule if needed)
    
    return df.apply(merge_row, axis=1)
```

```python
# Specify the columns to merge
columns_to_merge = ['Lab_Frequency_1097347','Lab_Frequency_1097346','Lab_Frequency_1087820']

# Create the merged column
data['Lab_frequency'] = merge_specified_columns(data, columns_to_merge)
```
